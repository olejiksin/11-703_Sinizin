```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as seab

file=r'../1task/train.csv'

def read(file):
    data= pd.read_csv(file)
    print(data.columns)
    print(data)
    return data

def plot(file):
    fullData=read(file)
    x1 = fullData.loc[fullData.Pclass==1]
    x2 = fullData.loc[fullData.Pclass==2]
    x3 = fullData.loc[fullData.Pclass==3]
    kwarks=dict(alpha=0.5,bins=20)
    plt.hist(x1['Age'],**kwarks,color='red',label='Pclass=1')
    plt.hist(x2['Age'],**kwarks,color='blue',label='Pclass=2')
    plt.hist(x3['Age'],**kwarks,color='green',label='Pclass=3')
    plt.gca().set(xlabel='Age',)
    plt.legend()
    
    graph=seab.FacetGrid(fullData, col='Pclass')
    graph.map(plt.hist, 'Age')
    return None;
    
    
plot(file)
```
