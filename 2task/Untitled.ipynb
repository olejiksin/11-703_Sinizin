```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file='../2task/Sunspots.csv'

def read(file):
    data= pd.read_csv(file)
    print(data.columns)
    return data

def draw(fil):
    data=read(fil)
    dateData=list(data.Date)
    sunspot=list(data['Monthly Mean Total Sunspot Number'])
    sunspotNew=[]
    sunspotNew.append(sunspot[0])
    alpha=0.8
    for i in range(1,len(sunspot)):
        sunspotNew.append(alpha*sunspot[i]+(1-alpha)*sunspot[i-1])
    plt.figure(figsize=(30,10))
    #plt.xlim(0,100)
    plt.scatter(dateData, sunspot)
    plt.scatter(dateData, sunspotNew,color='b')
    plt.plot(dateData, sunspot,color='g')
    plt.show()
    
draw(file)
```
